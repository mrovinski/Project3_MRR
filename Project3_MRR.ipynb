**Problem A:** Write a function that will accept the filename of data about African country population for 5 years and return the dataset as a list. 

```python
import csv

def readfile(filename):
    """This function reads a file and prints it into a list. It also replaces the \n at the end of all the lists with a blank. Because csv assumes that every comma is a separate element, the countries with commas in their names caused the length of each row to need to be taken into account to be corrected for output."""
    ans = []
    with open (filename, mode = 'r') as file:
        for rec in file:
            l = rec.replace('\n','').split(',')
            if len(l) > 7:
                l = [l[0] + ',' + l[1]] + l[2:]
            ans.append(l)
    return ans

data = readfile('africa_pop_data.csv')
print(data)
```

**Problem B:** Write a function that will accept a dataset (list) that contains the African country populations that will return a NumPy array with only the numerical data. 

```python
import numpy as np
```

```python
def pop_array(setlist):
    """This function takes the data from a list and puts it into a NumPy Array. With the length of each countries data being different, it had to call the last 5 elements for consistency. So it takes each sublist within the main list except the first one and uses the last five elements to make each row in the Array. """
    arr = [sub[-5:] for sub in setlist[1:]]
    for x in range(len(arr)):
        for i in range(len(arr[x])):
            arr[x][i] = int(arr[x][i])
    return np.array(arr)
nparr = pop_array(data)
print(nparr)
```

**Problem C:** Write a function that will accept a data list that contains the African country populations that will return PrettyTable of the data. 

```python
from prettytable import PrettyTable    
```

```python
def pop_pt(datas):
    """This function creates a PrettyTable of the Data from the data list. It makes the elements in the sublist 0 and makes it the headers. Then it takes all the data in sublist 1 through the end and inputs the data in each row."""
    pt = PrettyTable()
    pt.field_names = datas[0]
    for row in datas[1:]:
        pt.add_row(row)
    print (pt)
pop_pt(data)
```

**Problem D:** Write a function that will accept the NumPy array that contains the population data for the African countries and that prints a summary table of statistics for each year.

```python
def helper(vals):
    """This function formats the numbers in the arrays to have two decimal places and commas. It takes every three digits in an element and places a comma in between them and then appends each three digit section together. It also cuts off the last two digits of the numbers at the end of the number and then appends them after a decimal place at the end (which serves a purpose in the following function when each mean, max, and min is multiplied by 100)."""
    res = []
    for val in vals:
        num = int((len(str(int(val)))-3)/3)
        ans = ''
        for i in range(num):
            ans = ',' + val[len(val)-3-3*i - 2:len(val)-3*i - 2] + ans
        ans = val[:len(val) - 3*num - 2] + ans + '.' + val[-2:]
        res.append(ans)
    return res

def pop_stat(data):
    """This function creates a pretty table for the summary output. It prints a title for the chart and each row with the mean, min, and max for each year. It multiplies each mean, min, and max by 100 and then converts them to integers to preserve the first two digits of the decimal as was mentioned in the previous function."""
    pt = PrettyTable()
    print ("African Population Statistics")
    pt.field_names = ['Summary Measure','2015', '2016', '2017', '2018', '2019']
    
    mean = np.mean(data, axis=0) * 100
    maxi = np.max(data, axis=0) * 100
    mini = np.min(data, axis=0) * 100
    means = []
    maxes = []
    mins = []
    for i in range(len(mean)):
        means.append(str(int(mean[i])))
        maxes.append(str(maxi[i]))
        mins.append(str(mini[i]))
    
    row = helper(means)
    pt.add_row(['mean'] + row)
    row = helper(maxes)
    pt.add_row(['max'] + row)
    row = helper(mins)
    pt.add_row(['min'] + row)
     
    pt.align = "r"
    print(pt)
pop_stat(nparr)
```
